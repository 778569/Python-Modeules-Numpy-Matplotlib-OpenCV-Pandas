```python
import numpy as np

x=np.arange(0,100,2)
print(x)

y1 = 2*x+3
y2 = np.power(x,2)
print(y1)
print(y2)


import matplotlib.pyplot as plt
plt.plot(x,y1,'r^',label='X vs Y1')
plt.plot(x,y2,'g+' ,label='X vs Y1')
plt.legend()
plt.xlabel('X_values')
plt.ylabel('Y_values')
plt.savefig('MyGraph.png',dpi=300)
plt.title('XvsY')

plt.show()
```
